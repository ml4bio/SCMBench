```python
import pandas as pd
import scanpy as sc
```

```python
# data_name='Chen-2019-small'
data_name='Muto-2021-small'
```

```python
data=pd.read_csv('/mnt/nas/user/yixuan/scFoundation/apiexample/data/baron_human_samp_19264_fromsaver_demo.csv')
```

```python
# rna=sc.read_h5ad('/mnt/nas/user/yixuan/Multiomics-benchmark-main/data/download/'+data_name+'/'+data_name+'-RNA.h5ad')
rna=sc.read_h5ad('/mnt/nas/user/yixuan/Multiomics-benchmark-main/data/download/'+data_name+'/'+data_name+'-ACTIVE.h5ad')
```

```python
rna_df=pd.DataFrame(rna.X.todense(),index=rna.obs_names,columns=rna.var_names)
```

```python
for i in list(rna_df.columns)[1:]:
    if i in data.columns:
        continue
    else:
        rna_df.drop(i,axis=1,inplace=True)
rna_df.shape
```

```python
for i in list(data.columns)[1:]:
    if i in rna_df.columns:
        continue
    else:
        rna_df[i]=0
rna_df.shape
```

```python
rna_df_sorted = rna_df.sort_index(axis=1)
rna_df_sorted
```

```python
# Determine the number of files needed
num_files = rna_df_sorted.shape[0] // 200 + 1
print(num_files)
# Slice the DataFrame and save each chunk to a separate file
for i in range(num_files):
    start_row = i * 200
    end_row = (i + 1) * 200
    sliced_df = rna_df_sorted.iloc[start_row:end_row, :]
    sliced_df.to_csv(f'/mnt/nas/user/yixuan/scFoundation/apiexample/data/act/muto/10x_file_{i+1}.csv')  # Save the chunk to a CSV file
    # sliced_df.to_csv(f'/mnt/nas/user/yixuan/scFoundation/apiexample/data/atac/muto3/10x_file_{i+1}.csv')  # Save the chunk to a CSV file
```
